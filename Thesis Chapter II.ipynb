# CHAPTER II: A MODEL FOR THE FUTURE OF THE MUTUALLY TIDAL LOCKED EARTH-MOON SYSTEM

In this chapter, we will discuss the model in full detail and then proceed derive the equations for our model for the perturbative Earth-Moon tidal locking.

## 2.1 THE MODEL EXPLAINED

In this section, we will explain the model part by part, the assumptions behind it, and its advantages and limitations. We will derive differential equations that would determine, with the given initial conditions of some factors to be discussed later, the time when the rotational period of the Earth will approach approximately 99.99% of the orbital period of the Moon, in which their initial values are 24 hours and 27.13 days respectively. In short, our model will predict how many years from now will the Earth-Moon system be mutually tidal locked; wherein only one side of the Earth will ever see the Moon. The mechanism that would be responsible for this would be the tidal friction caused by the tidal bulges attempting to catch up with the Moon, which in turn produces torque.

### 2.1.1 ASSUMPTIONS OF THE MODEL

Usually when it comes to creating mathematical models, we have to consider and separate among all factors, what we should include fundamentally from what we could consider negligible enough to be removed. This is commonly done by physicists so that solving a particular problem would be considerably simplified without having to sacrifice
its accuracy by a lot of magnitudes. After all, depending on the objectives that need to be accomplished, the goal is to produce a model with minimum effort but maximum results.


In the case of our model, we will assume that the Earth is a perfectly spherically symmetric ball of rock covered by an ocean of uniform depth. The shape of the ocean becomes a hollow prolate spheroid once interacting tidally with the Moon and its volume is globally conserved, therefore making the inferred former shape of the ocean as a hollow sphere as a reference instead. The orbit of the Moon would be circular and its orbital plane will be perpendicular to the Earth’s axis, therefore libration does not occur. The gravitational effects of the Sun would be considered negligible. Since the source of tidal friction in our model involves the tidal bulges travelling along the solid Earth’s surface, we have to assign the dimensions for the bulges that involves not just the part of the ocean that bulges away from the normal sea level as commonly depicted in most illustrations, but also the part of the ocean under it. For the value of the high and low tides, we select the daily values obtained from Poipu, Hawaii, during a five-day period (December 25-31, 2013), then averaged. The levels there are very similar to the conditions of our assumed model of the Earth.

### 2.1.2 ADVANTAGES OF THE MODEL

The main advantage of this model is its simplicity wherein we could model the evolution of the Earth-Moon tidal system by using only the most basic of physical laws. Through our previously mentioned assumptions, only the most common computers of this era are needed to solve the differential equations, therefore eliminating the need for highly advanced and expensive supercomputers. This further solidifies our model as a very
practical tool that could even be applied to model the tidal interactions of other celestial objects that are already tidally locked at their initial conditions.

### 2.1.3 LIMITATIONS OF THE MODEL

The simplicity of the model will give it several limitations. A lot of factors have to be eliminated that could have initially minimal but long-term influences, such as the tilt of Earth’s axis, the actual orbital elements of the Moon’s orbit, the gravitational influence of the Sun which may include its own stellar evolution, the shape of Earth’s islands and continents, the elasticity of the Earth’s crust and therefore the effects of the solid Earth tides, the layers of the Earth, the fluid properties of the ocean and other counterintuitive factors. Atmospheric and geophysical factors are also ignored.


The most severe limitation of this model is that the ocean and the bulges were treated as non-fluids, therefore there is no proper coefficient of friction between the bulges and the solid Earth. The harmonic oscillations of the ocean due to tides were also not taken into account.

In this section, we derive the non-linear ordinary differential equation for the evolution of Earth’s rotational velocity over time. We will use a simplified model where the Earth is assumed to have a completely flattened land surface covered by an ocean with a uniform depth while the Moon is orbiting the equatorial plane of the Earth.

In our model, the Moon will raise two tidal bulges. Since the angular velocity of the Earth is greater than that of the Moon, there will be friction between the tidal bulges and the Earth’s surface. This friction will do work that will cause a decrease in the Earth’s rotational kinetic energy via the work-energy theorem for the non-conservative system,

\begin{align}
dW = dT_{E}
\end{align}

where dW is the infinitesimal work by friction and dTE infinitesimal change in kinetic energy. The infinitesimal work done by friction is given by,

\begin{align}
\delta W = -F_{F} ds
\end{align}

Where Ff is the frictional force done over the arc ds. The negative sign comes from the dissipative nature of the system. Since the frictional force is directly proportional to the coefficient of friction μk and normal force NB, we can rewrite equation (2) as equation (3);

# Constants:

$\Omega $ = initial Ω rotation of earth in radians/years, with respect to distant stars

$\omega $ = initial ω revolution of moon in radians/secs

$M$ = mass of Earth

$R$ = radius of Earth

$N_{B}$ = Normal force

$r$ = radius of moon's orbit

$m$ = mass of Moon

$\eta$ = percentage of energy lost by the earth due to tidal friction that goes to the moon

$\mu$ = coefficient of friction

$G$ = Gravitational constant in SI units

$\gamma$ = product of G and M; known as the Standard Gravitational Parameter


The following constants were coded as the following: 

$\Omega $ as Om0, $\omega $ as om0, $M$ as M, $R$ as R, $N_{B}$ as N_B, $r$ as r, $m$ as m, $\eta$ as eta, $\mu$ as u ,$G$ as G, and $\gamma$ as Y

# The Equations:

### Earth's Rotational Velocity, Ω(t)

\begin{align}
-\mu_{k}N_{B}(\Omega-\omega)=\frac{2}{5}MR_{E}\Omega\bigg(\frac{d\Omega}{dt}\bigg)
\end{align}

### The Moon's Angular Velocity, ω(t)

\begin{align}
\frac{6}{5}\eta MR^{2}_{E}\Omega\bigg(\frac{d\Omega}{dt}\bigg)=\gamma^{\frac{2}{3}}m\omega^{-\frac{1}{3}}\bigg(\frac{d\omega}{dt}\bigg)
\end{align}

### Earth-Moon Distance, r(t)

\begin{align}
r=\gamma^{\frac{1}{3}}\omega^{-\frac{2}{3}}
\end{align}

# Solving for Ω(t) and ω(t):


$\big(\frac{d\Omega}{dt}\big)=-\frac{5}{2}\frac{\mu_{k}N_{B}(\Omega-\omega)}{MR_{E}\Omega},\big(\frac{d\omega}{dt}\big)=\frac{6}{5}\frac{\eta MR^{2}_{E}\Omega}{\gamma^{\frac{2}{3}}m\omega^{-\frac{1}{3}}}\big(\frac{d\Omega}{dt}\big)$


```python
Om0=2301.21649
om0=84.889944
M=5.972*(10**24.0)
R=6371000.0
NB=1.02992455*(10**37) 
r=1737400.0  
m=7.34767309*(10**22.0)
eta=1.0/30.0 
u=0.5731*10**(-12) 
G=66425.34 
Y = G*M
#d0 = 384400000

def W_dt(W, t):
    return [(-5*u*NB*(W[0]-W[1]))/(2*M*R*W[0]), 
            ((6*eta*M*(R**2)*W[0])/(5*Y**(2/3)*m*W[1]**(-1/3)))*(-5*u*NB*(W[0]-W[1]))/(2*M*R*W[0])]

ts = np.arange(0,10*10**10,10**7)
W0 = [Om0, om0]
Ws = odeint(W_dt, W0, ts)
Earth = np.array(Ws[:,0])
Moon = np.array(Ws[:,1])
```

```python
plt.style.use('seaborn-whitegrid')
plt.plot(ts, Earth, label="Earth", c='blue')
plt.plot(ts, Moon, label="Moon", c='gray')
plt.xlabel("Time(years)")
plt.ylabel("Angular velocity (rad/yr)")
plt.title('Time Evolution of $\Omega$(t) when $\mu$$_k$=0.5731 x 10$^{-12}$')
plt.legend()
```

# Solving for r(t):


$r=\gamma^{\frac{1}{3}}\omega^{-\frac{2}{3}}$


```python
r=np.array((Y**(1/3))*Moon**(-2/3))
plt.plot(ts, r, c='green')
plt.xlim(0,ts[np.where(Moon/Earth < 0.99)][-1])
plt.xlabel("Time(years)")
plt.ylabel("Distance(m)")
plt.title('Earth-Moon Distance, r(t)')
```
